```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.metrics import r2_score,mean_squared_error
```

```python
temperature = pd.read_csv('temperature.csv')
temperature
```

```python
temperature.drop(['JAN-FEB','MAR-MAY','JUN-SEP','OCT-DEC'],axis = 1,inplace = True)
```

```python
temperature
```

```python
X = temperature[['YEAR']]
y = temperature['ANNUAL']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=42,shuffle = True)
```

```python
TemperatureModel = LinearRegression()
TemperatureModel.fit(X_train, y_train)
```

```python
y_pred = TemperatureModel.predict(X_test)
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse) 
print(f" mae : {mean_absolute_error(y_test,y_pred)}\n rmse : {rmse}\nscore : {r2_score(y_test,y_pred)}")
```

```python
res = TemperatureModel.predict([[2010],[2030],[2050]])
res
```

```python
CropYield = pd.read_csv('yield_df.csv')
```

```python
CropYield.head(3)
```

```python
IND = CropYield[CropYield['Area']=='India']
```

```python
IND.head(3)
```

```python
IND.reset_index(inplace = True)
```

```python
IND = IND.drop(['index','Unnamed: 0','Area','pesticides_tonnes'],axis = 1 )
```

```python
IND.head(3)
```

```python
IND['Item'].unique()
```

```python
Index = IND[(IND['Item'] == 'Cassava') | (IND['Item'] == 'Sorghum') | (IND['Item'] == 'Soybeans') | (IND['Item'] == 'Sweet potatoes')].index
```

```python
IND.drop(Index,inplace = True)
```

```python
IND.head(3)
```

```python
dummies = pd.get_dummies(IND.Item)
```

```python
merged = pd.concat([IND,dummies],axis = 1)
```

```python
merged.drop('Item',axis = 1,inplace = True)
```

```python
merged.head()
```

```python
merged.duplicated().sum()
```

```python
merged.drop_duplicates(inplace = True)
```

```python
merged.reset_index(inplace = True)
```

```python
merged.drop('index',axis = 1,inplace = True)
```

```python
merged.head(3)
```

```python
merged.describe()
```

```python
maize = merged.loc[merged['Maize']==True].drop(['Maize','Potatoes','Rice, paddy','Wheat'],axis = 1)
```

```python
maize.head(3)
```

```python
col = ['Year','average_rain_fall_mm_per_year','avg_temp','hg/ha_yield']
```

```python
maize = maize[col]
```

```python
maize.head(3)
```

```python
maize.describe()
```

```python
X = maize.iloc[:, :-1]
y = maize.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42, shuffle=True)
```

```python
MaizePredictor = LinearRegression()
MaizePredictor.fit(X_train,y_train)
```

```python
y_pred = MaizePredictor.predict(X_test)
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse) 
print(f" mae : {mean_absolute_error(y_test,y_pred)}\n rmse : {rmse}\nscore : {r2_score(y_test,y_pred)}")
```

```python
res = MaizePredictor.predict([[2050,1000,24],[2033,1100,25.5],[2029,1140,26]])
res
```
